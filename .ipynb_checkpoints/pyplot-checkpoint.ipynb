## 1. Introduction

This notebook contains an overview of the ```matplotlib.pyplot``` Python package and an in-depth explanation of three interesting plots from the package.

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python (https://matplotlib.org/stable/index.html) and can be used to create many different types of plot such as scatter plots and histograms. Matplotlin was originally inspired to emulate commands from MATLAB software (https://realpython.com/python-matplotlib-guide/) and ```matplotlib.pyplot``` is a collection of functions to make matplotlib work like MATLAB. MATLAB is a programming and numeric computing platform used to analyze data, develop algorithms, and create models (https://uk.mathworks.com/products/matlab.html).

A pyplot plot consists of a hierarchy of nested Python objects with a tree-like structure of matplotlib objects underlying each plot . The pylot hierarchy consists of a ```figure``` object which contains ```axes```. Axes in this context refers to an individual plot within the figure, as opposed to the mathmatical term meaning more than one axis. Below the ```axes``` in the hierarchy are smaller items such as individual lines, legends, text boxes, or plot titles. Most elements of a plot are Python objects which can be manipulated(https://realpython.com/python-matplotlib-guide/).

![realpython.com](https://files.realpython.com/media/fig_map.bc8c7cabd823.png)

Individual pyplot functions can be used to make changes to the figure. For example, creating a figure, creating a plotting area in a figure, or labelling a figure (https://matplotlib.org/stable/tutorials/introductory/pyplot.html).

### 3.1. Scatter Plot

The first interesting plot to be discussed is a scatter plot as detailed on the w3schools website: https://www.w3schools.com/python/matplotlib_scatter.asp. The plot has XXXXXXXXXXXXXXXXXXXXXXXXX:

The ```plt.scatter()``` function can be used to draw a scatter plot. It requires two arrays of equal length for the x- and y-axis (https://www.w3schools.com/python/matplotlib_scatter.asp). 

The full code for the plot is as follows. Please note, a seed is used for reproducability:

````python
np.random.seed(19680801) #seed added to code for reproducability
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
````

A breakdown and explanation of the code is as follows, using the seed for reproducability:

````python
np.random.seed(19680801)
````

The first two lines of code use the numpy random function to generate two arrays, one for the x-axis, and one for the y-axis. Each array contains 100 integers between 0 and 99; the integers can be repeated within each array.

````python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
x,y 
````

Each value on the x-axis corresponds to a value on the y-axis, in order as they appear.

The third line defines the variable colors. As with the previous two lines, an array of 100 random integers between 0 and 99 is created: 

````python
colors = np.random.randint(100, size=(100))
colors
````

The colors variable is used later in the code when calling the pyplot.scatter function to display different colours for different markers. Each value in the array corresponds with the values in the x-axis and y-axis, in order as the appear. 

The fourth line defines the variable sizes. As previously, an array of 100 random integers between 0 and 99 is created. Each integer that is generated is then multiplied by 10. 

````python
sizes = 10 * np.random.randint(100, size=(100))
sizes
````

The sizes variable is used later in the code when calling the pyplot.scatter function to display different sized markers. Each value in the array corresponds with the values in the x-axis and y-axis, in order as the appear. Each integer is multiplied by 10 to make the markers larger than they would otherwise appear. 

The final three lines of code:

1. Plot the scatter plot of y vs. x and set the paramaters c, s, alpha, and cmap.
2. Add a colorbar
3. Plot the complete figure

````python
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
````

1. The plt.scatter() function uses the integers from the x and y variables to plot the markets. The cmap parameter defines the colormap that the plot will use. In this instance, the colormap is nipy_spectral. A colormap is like a list of colors, where each color has a value that ranges from 0 to 100. There are a large number of built-in colormaps in the matplotlib package (https://www.w3schools.com/python/matplotlib_scatter.asp). The c parameter assigns a colour to each market from the 100 integers in the colors array. The alpha parameter is used to adjust the transparency of the markers. The value can be set between 0 (transparent) and 1 (opaque) (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html). The sizes parameter sets the marker size. 


2. A colorbar is added to the right of the plot, based upon the colormap. 


3. The complete figure is plotted.

An example of the above code but with different values set for x, y, colors, sizes, alpha, and with a different colormap is shown below: 

````python
x = np.random.randint(50, size=(50))
y = np.random.randint(50, size=(50))
colors = np.random.randint(50, size=(50))
sizes = 5 * np.random.randint(50, size=(50))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.75, cmap='magma')

plt.colorbar()

plt.show()
````

### 3.2. Stacked Area Plot

The second interesting plot to be discussed is a stacked area plot as detailed on the RealPython website: https://realpython.com/python-matplotlib-guide/. The plot is entitled "Combined debt growth over time" and represents the debt acumulated by three entities (Eastasia, Eurasia, and Oceania) over a period of time from 1950 to 1999. The debt for each entity is plotted as a stacked plot. 

The full code for the plot and the plot itself is as follows. Please note, a seed is used for reproducability:


````python
np.random.seed(19680801) #seed added to code for purposes of reproducability
rng = np.arange(50)
rnd = np.random.randint(0, 10, size=(3, rng.size))

yrs = 1950 + rng

fig, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
ax.set_title('Combined debt growth over time')
ax.legend(loc='upper left')
ax.set_ylabel('Total debt')
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
fig.tight_layout()
````

A breakdown and explanation of the code is as follows, using the seed for reproducability:

````python
np.random.seed(19680801)
````

The first line of code produces a range of 50 integers from 0 to 49:

````python
rng = np.arange(50)
rng
````

The second line of code produces 3 arrays which each contain 50 random integers from 0 to 10: 

````python
rnd = np.random.randint(0, 10, size=(3, rng.size))
rnd
````

Changing the value of the rng variable in the first line of code will change the number of integers in each array.

The third line of code creates an array of integers which reflect the years 1950 to 1999. Each of the integers in the rng array (from 0 to 49) is added to 1950:

````python
yrs = 1950 + rng
yrs
````

The next line calls the function ```plt.subplots()``` which returns the figure and axes objects (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html):

````python
fig, ax = plt.subplots(figsize=(5, 3))
````

As previously detailed, the pylot hierarchy consists of a figure object which contains axes; these objects are unpacked into the variables ``` fig```  and ``` ax``` (https://stackoverflow.com/questions/34162443/why-do-many-examples-use-fig-ax-plt-subplots-in-matplotlib-pyplot-python). The ```figsize``` keyword argument sets the width and height of the figure (https://www.southampton.ac.uk/~feeg1001/notebooks/Matplotlib.html).

The next line specifies the next level down in the hierarchy - the axes:

````python
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
````

```stackplot``` is used to draw a stacked area plot. The ```yrs```variable is used for the x-axis and represents the years 1950 to 1999. The y-axis is determined by each integer from 0 to 49 in ```rng``` being added to each of the corresponding random integers in ```rnd```; this is repeated for each of the three arrays in ```rnd```. ```stackplot``` Each part is given a label, and the the plot shows each part stacked onto one another (https://www.geeksforgeeks.org/matplotlib-pyplot-stackplot-in-python/).

At this point in the code, the data could be displayed:

````python
fig, ax = plt.subplots(figsize=(5, 3))
ax.stackplot(yrs, rng + rnd, labels=['Eastasia', 'Eurasia', 'Oceania'])
fig.tight_layout()
````

The next line of code sets the title for the plot:

````python
ax.set_title('Combined debt growth over time')
````

The next line of code sets the location for the legend within the plot. In this case, in the upper left corner.

````python
ax.legend(loc='upper left')
````

The next line of code sets the y-axis label:

````python
ax.set_ylabel('Total debt')
````

The next line of code sets the y-axis label between 1950 and 1999:

````python
ax.set_xlim(xmin=yrs[0], xmax=yrs[-1])
````

The first or minimum value along the x-axis is set by slicing the first element of the array:

````python
yrs[0]
````

The second or maximum value along the y-axis is set by slicing the last element of the array:

````python
yrs[-1]
````

The final line of code is used to ensure that the axis labels and titles do not go outside the figure area. In pyplot, the location of axes including subplots are specified in normalized figure coordinates and can sometimes go outside the figure area and become clipped. The fig.tight_layout() method can automatically account for this (https://www.pythonpool.com/matplotlib-tightlayout/).

````python
fig.tight_layout()
````

### 3.3. Bar plot

The third interesting plot to be discussed is a bar plot as detailed on the Scipy Lectures website: http://scipy-lectures.org/intro/matplotlib/auto_examples/plot_bar.html#sphx-glr-intro-matplotlib-auto-examples-plot-bar-py. The plot represents.....

The debt for each entity is plotted as a stacked plot.

The full code for the plot and the plot itself is as follows. Please note, seeds are used for reproducability:

````python
n = 12
X = np.arange(n)
np.random.seed(12345) #seed added to code for purposes of reproducability
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
np.random.seed(54321) #seed added to code for purposes of reproducability
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va= 'bottom')

for x, y in zip(X, Y2):
    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va= 'top')

plt.xlim(-.5, n)
plt.xticks([])
plt.ylim(-1.25, 1.25)
plt.yticks([])

plt.show()
````

The first line of code defines a variable n:

````python
n = 12
n
````

The second line of code creates the array X with a range of integers from from 0 to 11, based on the variable n. This defines the number of bars on the x-axis:

````python
X = np.arange(n)
X
````

The third line of code defines a variable Y1 which contains 12 numbers (based on the variable n). Y1 represents the first y-axis which rises vertically from 0 on the axis.

````python
np.random.seed(12345)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
````

This is broken down as follows. The numpy random function is used to generate an array of 12 with a minimum value of 0.5 (inclusive) and a maximum of 1.0 (exclusive):

````python
np.random.seed(12345)
a = np.random.uniform(0.5, 1.0, n)
a
````

Each number within this array is then multiplied with (1 - X / n). For example, for the first number in the array:

````python
np.random.seed(12345)
b = (1 - 0 / 12) * a[0] 
b
````

And for the twelth number:

````python
np.random.seed(12345)
c = (1 - 11 / 12) * a[-1]
c
````

These correspond to the values in Y1 on the first y-axis (rounded):

````python
Y1
````

The Y2 variable represents the second y-axis which descends vertically from 0 on the axis. The formula is repeated as in Y1.

````python
np.random.seed(54321)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2
````

The data used to make up the bar plot has now all been defined. 

The next line of code uses the axes function to take a list of [xmin, xmax, ymin, ymax] and specifies the viewport of the axes (https://matplotlib.org/stable/tutorials/introductory/pyplot.html):

````python
plt.axes([0.025, 0.025, 0.95, 0.95])
````

The next line plots the bar plot using along the x-axis and the first y-axis. The + symbol is used to show the y-axis data rising vertically. A hexidecimal colour is chosen for the bars with an edgecolour of white:

````python
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
````

The next line plots the bar plot using along the x-axis and the second y-axis. The - symbol is used to show the y-axis data descending vertically. A different hexidecimal colour is chosen for the bars with an edgecolour of white:


````python
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')
````

The next line of code add labels to the axes, to remove the ticks, and set the x- and y-axis limits. 

First, for loops add labels to the Y1 and Y2 axes representing the corresponding Y1 and Y2 values. The code loops through each axes to set the positioning of the labels - slightly right of the x value, and slightly above the y. The Y1 and Y2 values are then rounded to 2 decimal figures. The horizontal alignment is set to centre, and the vertical alignment to bottom for the Y1-axis and to top for the Y2-axis.

````python
for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va= 'bottom')
    
for x, y in zip(X, Y2):
    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va= 'top')
    
plt.xlim(-.5, n)
plt.xticks([])
plt.ylim(-1.25, 1.25)
plt.yticks([])

````
